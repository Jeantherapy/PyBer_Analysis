# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
rides_by_city_type = pyber_data_df.groupby('type')['ride_id'].count()
rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
drivers_by_city_type = city_data_df.groupby('type')['driver_count'].sum()
drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
# Group the data by city type and calculate the sum of fares
fares_by_city_type = pyber_data_df.groupby('type')['fare'].sum()

# Display the total amount of fares for each city type
fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
# Calculate the average fare per ride for each city type
average_fare_per_ride = fares_by_city_type / rides_by_city_type

# Display the average fare per ride for each city type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = fares_by_city_type / drivers_by_city_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
# Create a dictionary with the summary data
summary_data = {
    'Total Rides': rides_by_city_type,
    'Total Drivers': drivers_by_city_type,
    'Total Fares': fares_by_city_type,
    'Average Fare per Ride': average_fare_per_ride,
    'Average Fare per Driver': average_fare_per_driver
}

# Create the PyBer summary DataFrame
pyber_summary_df = pd.DataFrame(summary_data)

# Display the PyBer summary DataFrame
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
# Format the columns in the PyBer summary DataFrame
pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map('{:,}'.format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map('{:,}'.format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map('${:.2f}'.format)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map('${:.2f}'.format)

# Display the formatted PyBer summary DataFrame
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
# Convert the 'date' column to datetime data type
pyber_data_df['date'] = pd.to_datetime(pyber_data_df['date'])

# Group the data by city type and date, and calculate the sum of fares
fare_sum_by_type_date = pyber_data_df.groupby(['type', 'date'])['fare'].sum()

# Create a new DataFrame from the grouped data
fare_summary_df = pd.DataFrame(fare_sum_by_type_date)

# Display the new DataFrame
fare_summary_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

fare_summary_df = fare_summary_df.reset_index()

# Display the DataFrame with reset index
fare_summary_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

pivot_table = pd.pivot_table(fare_summary_df, index='date', columns='type', values='fare')

# Display the pivot table
pivot_table
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
# Filter the pivot table DataFrame for the specified dates
filtered_df = pivot_table.loc['2019-01-01':'2019-04-28']

# Display the new DataFrame
filtered_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
#df.index = pd.to_datetime(df.index)
# Set the "date" index to datetime datatype
filtered_df.index = pd.to_datetime(filtered_df.index)

# Display the DataFrame with the updated index datatype
filtered_df
```

```python
# 6. Check that the datatype for the index is datetime using df.info()

filtered_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
# Resample the DataFrame by week and calculate the sum of fares
weekly_fares = filtered_df.resample('W').sum()

# Display the new DataFrame
weekly_fares
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Import the style from Matplotlib
from matplotlib import style

# Use the graph style fivethirtyeight
style.use('fivethirtyeight')

# Plot the resampled DataFrame using the object-oriented interface
ax = weekly_fares.plot(figsize=(18,5))

# Set the plot title and axis labels
ax.set_title('Weekly Fares')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')

# Display the plot
plt.show()
plt.savefig("analysis/PyBer_fare_summary.png")
```
